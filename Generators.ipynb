```python
def gen_demo():
  yield "first"
  yield "second"
  yield "third"
```

```python
gen = gen_demo()
print(gen)   # returns generator objects
```

```python
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
for i in gen_demo():
  print(i)
```

```python
def square(num):
  for i in range(1,num+1):
    yield i**2
```

```python
gen1 = square(10)
print(next(gen1))
print(next(gen1))
for i in gen1:
  print(i)
```

```python
print(next(gen1)) # error caues we have already calledd all the elements
```

```python
# range function using generator
def my_range(num):
  for i in range(num):
    yield i
```

```python
num = my_range(100)
for i in num:
  print(i)
```

```python
gen3 = (i for i in range(10))
for i in gen3:
  print(i)
```

 ## Chaining Generators

```python
def fibo(nums):
  x,y =0,1
  for i in range(nums):
    x,y=y,y+1
    yield x

def square(nums):
  for num in nums:
    yield num**2

```

```python
print(sum(square(fibo(10))))
```
