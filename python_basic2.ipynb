```python
class Person:
  def __init__(self, name, lname, age, gender):
    self.name = name
    self.lname = lname
    self.age = age
    self.gender = gender
```

```python
person = Person('biswas','pokhrel',21, 'male')

def show_object(person):
  if isinstance(person, Person):
    return {
        'name': person.name,
        'lname':person.lname,
        'age': person.age,
        'gender':person.gender
    }
```

```python
import json
with open('sample.json','w') as f:
  json.dump(person,f,default=show_object,indent=4)
```

```python
with open('sample.json','r') as f:
  print(type(json.load(f)))
```

```python
import pickle
with open('sample.pkl', 'wb') as f:
  pickle.dump(person,f)
```

```python
def decorators(func):
  def wrapper():
    print("*"*10)
    func()
    print("*"*10)
  return wrapper

@decorators
def hello():
  print("Hello world")

hello()
```

```python
import time

def timer(func):
    def wrapper(*args):
      start = time.time()
      func(*args)
      print("tike taken:",func.__name__,time.time()-start)
    return wrapper

@timer
def square(num):
  return num**2

square(2)

```

```python
def check(data_types):
  def outer_wrapper(func):
    def inner_wrapper(*args):
      if type(*args) == data_types:
        func(*args)
      else:
        raise TypeError("not a valid datatypes")
    return inner_wrapper
  return outer_wrapper

@check(int)
def square(num):
  print(num**2)

square(2)
```

```python
def multtyply(a,b):
  mul = 0
  for i in range(b):
    mul+=a
  return mul
```

```python
def mul(a,b):
  if b==1:
    return a
  else:
    return a + mul(a,b-1)
```

```python
print(mul(3,4))
```

```python
def fact(number):
  if number == 1:
    return 1
  else:
    return number*fact(number-1)
```

```python
print(fact(3))
```

```python
# palindrome using recursion
def palindrome(string):
  if len(string) <= 1:
    return "palindrome"
  else:
    if string[0] == string[-1]:
      return palindrome(string[1:-1])
    else:
      return "not a palindrome"
```

```python
print(palindrome("mada"))
```

```python
print(palindrome("abba"))
```

```python
# Rabit problem
def fibo(mon):
  if mon ==1 or mon ==0:
    return 1
  else:
    return fibo(mon-1) + fibo(mon-2)
```

```python
import time
start = time.time()
print(fibo(45))
print(time.time()-start)
```

```python
def memo(n,d):
  if n in d:
    return d[n]
  else:
    d[n] = memo(n-1, d) + memo(n-2, d)
    return d[n]

d ={0:1, 1:1}
start1 = time.time()
print(memo(48,d))
print(time.time()-start1)
```

```python
# own for loop
def loop(iterable):
  iterator = iter(iterable)
  while True:
    try:
      print(next(iterator))
    except StopIteration:
      break
```

```python
l = [1,2,3,4,5,6,7]
loop(l)
```

```python
class my_range:
  def __init__(self, start, end):
    self.start = start
    self.end =end

  def __iter__(self):
    return range_iterator(self)


class range_iterator:
  def __init__(self, iterable_obj):
    self.iterable = iterable_obj

  def __iter__(self):
    return self

  def __next__(self):
    if self.iterable.start >=self.iterable.end:
      raise StopIteration
    current = self.iterable.start
    self.iterable.start +=1
    return current


```

```python
for i in my_range(1,100):
  print(i)
```
